```python
import pandas as pd
from IPython.display import display,HTML
from bs4 import BeautifulSoup
import requests
import re

urls = [] # list of site urls

# recieving urls with page=n structure
page = 1  # start page
while page != 30: # last page of search 
    urls.append(f"https://www.dominomusic.com/releases?page={page}")  
    page = page + 1

# collecting urls of 300x300 album covers   

album_cover = [] # list of album covers urls
for url in urls:
    for row in BeautifulSoup(requests.get(url).content, 'html.parser').find_all(
        'img', attrs={"class": "img-responsive-full transition lazyload", "alt":"", "data-sizes":"auto"}
):
        album_cover.append(row['data-src'])

artist_name = []
for url in urls:
    for row in BeautifulSoup(requests.get(url).content, 'html.parser').find_all(
    'span', attrs={'class': 'artist-name block'}
):
         artist_name.append(row.text)
            
release_name = []  # release list
for url in urls:
    for row in BeautifulSoup(requests.get(url).content, 'html.parser').find_all(
    'span', attrs={'class': 'release-title block'}
):
        release_name.append(row.text)  


release_date = []  # release date
for url in urls:
    for row in BeautifulSoup(requests.get(url).content, 'html.parser').find_all(
    'span', attrs={'class': 'd-none d-md-inline-block'}
):
        release_date.append(row.text)
            
products_data = pd.DataFrame() 
products_data['artist_name'] = artist_name
products_data['release_name'] = release_name
products_data['release_date'] = release_date
products_data['release_date'] = pd.to_datetime(products_data['release_date'])
products_data['album_cover'] = album_cover

def path_to_image_html(path):
    return '<img src="'+ path + '" width="60">' # choose size of image

pd.set_option('display.max_colwidth', None)

image_cols = ['album_cover']  #define which columns will be used to convert to html


format_dict = {} # setting the dictionariy 
for image_col in image_cols:
    format_dict[image_col] = path_to_image_html


# display(HTML(products_data.to_html(escape=False ,formatters=format_dict)))
```

```python
display(HTML(products_data.query('artist_name == "Franz Ferdinand"').to_html(escape=False ,formatters=format_dict)))
```
